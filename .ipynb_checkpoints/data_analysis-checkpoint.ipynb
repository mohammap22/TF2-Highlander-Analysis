```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tf2_match_data/tf2_stats_32000_matches.csv')

# Define weights for class-specific frags
class_weights = {
    'medic_frags': 3,  # Highest weight for Medic
    'demoman_frags': 2,  # High weight for Demoman and Sniper
    'sniper_frags': 2,
    'heavyweapons_frags': 1.5,
    'pyro_frags': 1, 'scout_frags': 1, 'soldier_frags': 1, 'engineer_frags': 1,
    'spy_frags': 0.75 # Lowest weight for Spy
}

# Fill NaN values in frags columns with 0
class_frags_columns = [col for col in df.columns if '_frags' in col]
df[class_frags_columns] = df[class_frags_columns].fillna(0)

# Calculate weighted class-specific frags sum
df['weighted_class_specific_frags_sum'] = sum(df[col] * weight for col, weight in class_weights.items())

# Function to calculate class_score
def calculate_base_class_score_updated(row):
  if row['player_class'] in ['undefined', 'unknown']: return None
 
  kills_weight = 1
  assists_weight = .5
  cpc_weight = 1  # 'cpc' for control point captures
  deaths_weight = -1  # Fewer deaths is better
  damage_weight = 1/1000  # 'dmg' for damage
  dt_weight = -0.5/1000 # 'dt' for damage taken, less impactful than damage output
  drops_weight = 4  # Drops, particularly bad for Medics

  match_minutes = row['match_time_seconds'] / 60
  # Calculate class score with corrected column names and considerations
  base_score = (row['assists'] * assists_weight +
                  row['cpc'] * cpc_weight +
                  row['deaths'] * deaths_weight +
                  row['dmg'] * damage_weight +
                  row['dt'] * dt_weight +
                  row['drops'] * drops_weight +
                  row['weighted_class_specific_frags_sum']) / match_minutes


  return base_score

def adjust_class_score_based_on_avg_performance(df):
    # Calculate the average base_class_score for each class
    avg_scores = df.groupby('player_class')['base_class_score'].mean()

    # Function to adjust score based on performance relative to the average
    def adjust_score(row, avg_scores):
        adjustment_factor = 1  # Initialize adjustment factor

        # For Pyro, Scout, Soldier, Heavy, and Sniper, check performance of enemy classes
        if row['player_class'] in ['pyro', 'scout', 'soldier', 'heavyweapons', 'sniper']:
            enemy_team = 'Blue' if row['team'] == 'Red' else 'Red'
            match_id = row['match_id']

            # Filter enemy team players from the same match
            enemy_df = df[(df['match_id'] == match_id) & (df['team'] == enemy_team)]

            # Conditions for "did poorly"
            if row['player_class'] == 'pyro' or row['player_class'] == 'heavyweapons':
                # Check if enemy Spy and Soldier did poorly
                enemy_spy_score = enemy_df[enemy_df['player_class'] == 'spy']['base_class_score'].mean()
                enemy_soldier_score = enemy_df[enemy_df['player_class'] == 'soldier']['base_class_score'].mean()

                if enemy_spy_score > avg_scores['spy'] and enemy_soldier_score > avg_scores['soldier']:
                    adjustment_factor -= 0.25
                if enemy_spy_score < avg_scores['spy'] and enemy_soldier_score < avg_scores['soldier']:
                    adjustment_factor += 0.25

            if row['player_class'] in ['scout', 'soldier']:
                # Check if enemy Scout and Soldier did poorly
                enemy_scout_score = enemy_df[enemy_df['player_class'] == 'scout']['base_class_score'].mean()
                enemy_soldier_score = enemy_df[enemy_df['player_class'] == 'soldier']['base_class_score'].mean()

                if enemy_scout_score > avg_scores['scout'] and enemy_soldier_score > avg_scores['soldier']:
                    adjustment_factor -= 0.25
                if enemy_scout_score < avg_scores['scout'] and enemy_soldier_score < avg_scores['soldier']:
                    adjustment_factor += 0.25

            if row['player_class'] == 'sniper':
                # Check if enemy Sniper did poorly
                enemy_sniper_score = enemy_df[enemy_df['player_class'] == 'sniper']['base_class_score'].mean()
                if enemy_sniper_score > avg_scores['sniper']:
                    adjustment_factor -= 0.25
                if enemy_sniper_score < avg_scores['sniper']:
                    adjustment_factor += 0.25

        if row['match_id'] % 500 == 0:
          print(row['match_id'])
        # Adjust the base_class_score by the calculated adjustment factor
        return row['base_class_score'] * adjustment_factor

    # Apply the adjustment logic to each row and create a new adjusted_class_score column
    df['class_score'] = df.apply(adjust_score, args=(avg_scores,), axis=1)
    return df


# Apply the class_score calculation to the entire dataframe
df['class_score'] = df.apply(calculate_base_class_score_updated, axis=1)
```

```python
# Filter out the 'unknown', 'undefined', and 'medic' classes before calculating averages
filtered_data = df[~df['player_class'].isin(['unknown', 'undefined', 'medic'])]

# Calculate the average class score for each class (excluding 'unknown', 'undefined', 'medic'), ignoring the map
avg_class_scores_exclusive = filtered_data.groupby('player_class')['class_score'].mean().sort_values().reset_index()

# Plotting the average class scores for each class
plt.figure(figsize=(10, 6))
sns.barplot(x="class_score", y="player_class", data=avg_class_scores_exclusive)
plt.title('Average Class Scores (Excluding Medic)')
plt.xlabel('Average Class Score')
plt.ylabel('Player Class')

plt.show()
```

```python
# Recalculate the average class score for each class on each map with the filtered data
avg_class_scores_filtered = filtered_data.groupby(['map', 'player_class'])['class_score'].mean().reset_index()

# Plotting the filtered data
plt.figure(figsize=(12, 8))
sns.barplot(x="class_score", y="map", hue="player_class", data=avg_class_scores_filtered)
plt.title('Average Class Score by Map (Excluding Medic)')
plt.xlabel('Average Class Score')
plt.ylabel('Map')
plt.legend(title='Player Class', bbox_to_anchor=(1.05, 1), loc='upper left')


plt.show()
```

```python
# Group by player_class and weapon_name, then count occurrences
weapon_counts_filtered = filtered_data.groupby(['player_class', 'weapon_name']).size().reset_index(name='counts')

# Sort within each player_class by counts to get the most common weapons on top
weapon_counts_filtered = weapon_counts_filtered.sort_values(['player_class', 'counts'], ascending=[True, False])

# Get the top 3 weapons for each class
top_weapons_per_class = weapon_counts_filtered.groupby('player_class').head(3)

# Calculate the average class_score for each class and weapon combination
avg_class_scores = filtered_data.groupby(['player_class', 'weapon_name'])['class_score'].mean().reset_index()

# Merge top weapons with their average class_score
top_weapons_scores = pd.merge(top_weapons_per_class, avg_class_scores, on=['player_class', 'weapon_name'], how='left')

# Sort values for better visualization
top_weapons_scores_sorted = top_weapons_scores.sort_values(['player_class', 'counts'], ascending=[True, False])

# Visualization
plt.figure(figsize=(20, 15))
classes = filtered_data['player_class'].unique()
for i, player_class in enumerate(sorted(classes)):
    plt.subplot(3, 3, i+1)
    class_data = top_weapons_scores_sorted[top_weapons_scores_sorted['player_class'] == player_class]
    bars = plt.bar(class_data['weapon_name'], class_data['class_score'], color='dodgerblue')
    plt.title(player_class.capitalize())
    plt.xticks(rotation=45, ha="right")
    plt.ylabel('Average Class Score')
    plt.xlabel('Weapon')
    plt.tight_layout()

plt.suptitle('Top 3 Weapons and Average Class Score by Player Class in TF2 Matches', fontsize=16, y=1.02)
plt.show()
```

```python
# Aggregate total kills and deaths for each class in each match
class_totals_per_match = filtered_data.groupby(['match_id', 'player_class']).agg(
    total_kills=pd.NamedAgg(column='kills', aggfunc='sum'),
    total_deaths=pd.NamedAgg(column='deaths', aggfunc='sum')
).reset_index()

# Calculate the K/D ratio for each class in each match
class_totals_per_match['kd_ratio'] = class_totals_per_match['total_kills'] / class_totals_per_match['total_deaths'].replace(0, 1)

# Calculate the average K/D ratio per class across all matches
average_kd_per_class = class_totals_per_match.groupby('player_class')['kd_ratio'].mean()

average_kd_per_class
```

```python
df_correct_composition = df[~df['player_class'].isin(['unknown', 'undefined'])]
df_correct_composition = df_correct_composition.drop(['weighted_class_specific_frags_sum', 'unknown_frags'], axis=1)

# Find all columns related to '*_frags'
frags_columns = [col for col in df_correct_composition.columns if '_frags' in col]


# Define the 9 specific TF2 classes again for clarity
valid_classes = ['scout', 'soldier', 'pyro', 'demoman', 'heavyweapons', 'engineer', 'medic', 'sniper', 'spy']

# Adjusting the approach to ensure we get a full breakdown of kills for each class against all others

# Displaying column names to understand the structure better and ensure correct aggregation
column_names = df_correct_composition.columns.tolist()

# Display column names related to class kills ('*_frags')
class_kill_columns = [col for col in column_names if '_frags' in col]

# For each class, we need to calculate kills against other classes
for player_class in valid_classes:
    # Initialize a dictionary to store kills by the current class on other classes
    kills_on_classes = {}

    for target_class in valid_classes:

        # Column name for kills on target class by player class
        column_name = f'{target_class}_frags'

        # Aggregate kills and store in the dictionary
        if column_name in class_kill_columns:
            d = df_correct_composition[df_correct_composition['player_class'] == player_class]
            kills_on_classes[target_class] = d[column_name].sum()

    # Convert dictionary to a Series for easier plotting
    kills_series = pd.Series(kills_on_classes).sort_values(ascending=False)

    # Plotting
    plt.figure(figsize=(10, 5))
    sns.barplot(x=kills_series.values, y=kills_series.index)
    plt.title(f'Most Common Classes Killed by {player_class.capitalize()}')
    plt.xlabel('Total Kills')
    plt.ylabel('Class Killed')
    plt.show()
```
